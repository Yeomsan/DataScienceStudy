```python
import os
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import dataloader, Dataset
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
transf = transforms.Compose([transforms.Resize((200, 200)),\
                             transforms.RandomHorizontalFlip(),\
                             transforms.ToTensor()])
```

```python
img = Image.open('image.png')

plt.imshow(img)
plt.axis("off")
plt.show()
print(np.array(img))
```

```python
pre_pro_img = transf(img)

plt.imshow(pre_pro_img.permute(1, 2, 0).numpy())
plt.axis("off")
plt.show()
```
